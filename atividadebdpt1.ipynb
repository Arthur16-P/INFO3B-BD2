```python
SELECT COUNT(*) AS total_grupos
FROM grupo_cultural;

SELECT COUNT(*) AS total_eventos
FROM evento;

SELECT grupo_cultural.nome,
       evento.local,
       evento.data
FROM grupo_cultural
JOIN apresenta
ON grupo_cultural.id = apresenta.grupo_id
JOIN evento
ON apresenta.evento_id = evento.id;

CREATE VIEW vw_eventos AS
SELECT id,
       data,
       horario,
       local,
       status
FROM evento;

DELIMITER //

CREATE PROCEDURE listarEventos()
BEGIN
    SELECT * FROM evento;
END //

DELIMITER ;

DELIMITER //

CREATE FUNCTION quantidadeEventos()
RETURNS INT
DETERMINISTIC
BEGIN
    DECLARE total INT;

    SELECT COUNT(*)
    INTO total
    FROM evento;

    RETURN total;
END //

DELIMITER ;

CREATE TRIGGER verificar_status
BEFORE UPDATE
ON evento
FOR EACH ROW
BEGIN
    IF NEW.status = '' THEN
        SET NEW.status = 'Pendente';
    END IF;
END;
```
